```python
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python

df['text'] = df['Subject'].fillna('') + ' ' + df['Message'].fillna('')
X = df['text']
y = df['Spam/Ham']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
vectorizer = TfidfVectorizer()
X_train_tfidf = vectorizer.fit_transform(X_train)
X_test_tfidf = vectorizer.transform(X_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier()
```

```python
rf.fit(X_train_tfidf, y_train)
```

```python
y_pred=rf.predict(X_test_tfidf,)
```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
print("Classification Report:")
print(classification_report(y_test, y_pred))
print("Confusion Matrix:")
print(confusion_matrix(y_test, y_pred))
```

```python
def predict_email(email_text):
    processed = vectorizer.transform([email_text])
    prediction = rf.predict(processed)  # replace with your model name
    result = "SPAM" if prediction[0] == 1 else "HAM"
    return result

user_email = input("Enter the email content:\n")
result = predict_email(user_email)
print(f"\nThe email is predicted to be: {result}")
```
